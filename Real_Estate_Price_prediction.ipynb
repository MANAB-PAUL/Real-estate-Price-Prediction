### spliting testing and training data

```python
import numpy as np
np.random.seed(42)
### function to split the data into test and train
def train_test_set(data, test_ratio = 0.2):
    indices = np.random.permutation(len(data))  ### returns a random permutation of array 0 to len(data) - 1
    test_size = int(test_ratio * len(data))
    
    test_indices = indices[:test_size]
    train_indices = indices[test_size:]
    
    return data.iloc[train_indices], data.iloc[test_indices]    
```

```python
train_set, test_set = train_test_set(housing)
```

# scikit Learn Design

### primarily three types of object
1. Estimators - estimates some parameter based on a dataset - eg - Imputer. It has fit and transform method.
1. Fit:- fits the dataset and calculates internal parameters
    
2. Transformers :- takes input and returns output based on lernings from fit().
    It also has a convinience function called fit_transform()
    
3. Predictors:- LinearRegression is an example of Predictors it has two common functions
    fit() and predict(). It also gives score function which
    will evaluate the predictions

## Feature Scaling

Primarily, two types of feature scalling methods:
1. Min-max scaling (Normalization)
    (value - min) / (max - min)
    sklearn provides a class for this called MinMaxScaler for this
    
   2. Standardization
       (value - mean) / std
       sklearn provides a class called Standard Scaler for this

```python
housing = train_set.drop('MEDV', axis = 1)
housing_labels = train_set['MEDV'].copy()
```

## Creating Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr
```

# Selecting a desired model for predicting Real Estates price

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(housing_num_tr, housing_labels)
```

```python
### evaluating the model
from sklearn.metrics import mean_squared_error
```

```python
housing_predictions = model.predict(housing_num_tr)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)

lin_rmse
```

```python
list(housing_labels)
```

```python
list(housing_predictions)
```

### decision tree regressor

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
### evaluating the model
model_predictions = model.predict(housing_num_tr)

lin_mse = mean_squared_error(housing_labels, model_predictions)
lin_rmse = np.sqrt(lin_mse)

lin_rmse
```

## Using better evaluation technique - Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, housing_labels, scoring = 'neg_mean_squared_error', cv = 10)
rmse_scores = np.sqrt(-scores)
```
